```python
data2.drop(['Order ID','Region','Product Category'],axis=1,inplace=True)
data2.columns
```

```python
# Step 1: Select only numeric columns
numeric_cols = data2.select_dtypes(include='number').columns

# Step 2: Calculate IQR only for numeric columns
Q1 = data2[numeric_cols].quantile(0.25)
Q3 = data2[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Step 3: Filter out the outliers based on IQR
outlier_filter = ~((data2[numeric_cols] < (Q1 - 1.5 * IQR)) | (data2[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

# Step 4: Apply the filter to the whole DataFrame (not just numeric columns)
data2 = data2[outlier_filter]

# Show result
data2.head()
```
